```python
import os
import torch
from torchvision.utils import save_image
from torch.utils.data import Dataset, DataLoader
from torch.utils.data.sampler import RandomSampler, BatchSampler

class Generator(object):
    def __init__(self, n_frames):
        self.n_frames = n_frames
    
    def generate(self, folder_path, prefix=""):
        front, back = torch.randn(2, self.n_frames, 3, 64, 64)
        for t in range(self.n_frames):
            front_img = f"img_{t}_front.png"
            back_img = f"img_{t}_back.png"
            if prefix:
                front_img = f"{prefix}_img_{t}_front.png"
                back_img = f"{prefix}_img_{t}_back.png"
            save_image(front[t], os.path.join(folder_path, front_img))
            save_image(back[t], os.path.join(folder_path, back_img))
```

```python
Generator(48).generate("./images/")
```

```python
class SynthDataset(Dataset):
    def __init__(self):
        self.data = torch.arange(100).view(100, 1).float()
        
    def __getitem__(self, index):
        x = self.data[index]
        print(x.shape)
        return x
    
    def __len__(self):
        return len(self.data)

dataset = SynthDataset()        
sampler = BatchSampler(
    RandomSampler(dataset),
    batch_size=10,
    drop_last=False)

loader = DataLoader(
    dataset,
    sampler=sampler)

for data in loader:
    print(data)
```
